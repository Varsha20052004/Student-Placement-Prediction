```python
import pandas as pd

# Load the dataset
file_path = '/content/grad.xlsx'
df = pd.read_excel(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
```

```python
# Parse the 'Created' column to extract the year
df['Created'] = pd.to_datetime(df['Created'])
df['Year Created'] = df['Created'].dt.year

# Display the updated dataframe
df.head()
```

```python
# Assuming the final academic year is 4
final_academic_year = 4

# Calculate the difference and add it to the 'Year Created' to get the 'Graduation Year'
df['Graduation Year'] = df['Year Created'] + (final_academic_year - df['Academic Year'])

# Display the updated dataframe
df.head()
```

```python
output_file_path = 'updated_grad.xlsx'

# Save the updated dataframe to a new Excel file
df.to_excel(output_file_path, index=False)
```
